```python
def Dataframe_from_LST20_Path(list_path):
    """return text file in directory to datframe
    input: folder_path which contains txt file
    output: dataframe"""
    line_list = []    
    for item in list_path:     
      # if item[0] != '.': # if have a '.' file in directory
        with open(item,'r') as f:
            for line in f:
                if line == '\n': #fill newline
                    continue                    
                line = line.strip().split('\t')          
                line_list.append(line)
                    
    df = pd.DataFrame(line_list,columns=['word','POS','name_entity-type','sentencs_seg']) #4 columns refer to 3 task of nlp
    #in case column NAN because newline, will be filled by some word
    df['POS'] = df['POS'].fillna("newline")
    df['name_entity-type'] = df['name_entity-type'].fillna("newline")
    df['sentencs_seg'] = df['sentencs_seg'].fillna("newline")
    return df
```

```python
import re
import time
import oskut
import pandas as pd
import numpy as np
```

```python
def text_to_special_word_idx(text, regex_pattern, verbose=True):
    if verbose == True:
        list_regex = [[m.start(0), m.end(0)] for m in re.finditer(regex_pattern, text)]
        word_dict = dict()
        for t in list_regex:
            word = text[t[0]:t[1]]
            if word_dict.get(word) is not None:
                word_dict[word].append([t[0],t[1]])
            else:
                word_dict[word] = [[t[0],t[1]]]
        return word_dict
    
    else:
        list_regex = [[m.start(0), m.end(0)] for m in re.finditer(regex_pattern, text)]
        return list_regex



def char_split_word_tag(word_list):
    df_ = pd.DataFrame(word_list, columns=['word'])

    word_tag = list()
    list_of_word = df_['word'].tolist()
    i = 0

    for word in list_of_word:
        isfirsttime = True
        index_last_char = word.rindex(word[-1])
        for idx, c in enumerate(word):
            if c == ' ':
                word_tag.append((i+1, c, 'O'))
            elif isfirsttime:
                word_tag.append((i+1,c, 'B_WORD'))
            elif idx == index_last_char:
                word_tag.append((i+1, c,'E_WORD'))
            else:
                word_tag.append((i+1, c,'I_WORD'))
            isfirsttime = False

            i += 1

    return pd.DataFrame(word_tag, columns=['Id', 'word','Predicted'])
```

```python
def create_binary_array(cut):
    binary_true = np.zeros(sum([len(w) for w in cut]))
    cum = 0
    for w in cut:
        binary_true[cum] = 1
        cum += len(w)
    return binary_true

def cut_word_from_bin(bin,cut_list,verbose=False):
    """
    Cut word from position pairs
    binary of cutting : len must equal to len(_txt)
    _txt : text before cut
    cut_list: can be either true of predict
    """
    txts = []
    txt_ = ''.join(cut_list)
    cut = np.where(bin==1)[0].tolist()
    if verbose:  print(cut)
    for i,c in enumerate(cut):
        # Last word
        # print(i)
        c = int(c)
        if i == len(cut) -1: 
            txts.append(txt_[c:])
        else:
            txts.append(txt_[c:int(cut[i+1])])
    return txts

def post_process_binary(pos_pair,pred,verbose=False):
    """
    Change binary from cut position pair
    pos_pair: list of postion pairs (starts,end)
    cut_list: predict cut list **Not binary
    return list of positons pairs [(starts,end),...]
    """
    txts = ''
    _txt = _txt = ''.join(pred)
    new_bin = create_binary_array(pred)
    for b,e in pos_pair:
        # if verbose : print,_txt[b:e])
        # before post process: 
        bef_post = cut_word_from_bin(new_bin[b:e],[_txt[b:e]])
        if verbose :  print('Before post',b,e,':',bef_post)

        # Process
        new_bin[b] = 1
        for i in range(b+1,e):
            new_bin[i] = 0
        if e+1 == len(new_bin):
            if new_bin[e+1] == 0: new_bin[e+1] =1 

        # after post process:
        aft_post = cut_word_from_bin(new_bin[b:e],[_txt[b:e]])
        if verbose :print('After post',aft_post)
    return new_bin
```
