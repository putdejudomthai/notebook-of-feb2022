# PyThaiNLP Transalate

```python
from pythainlp.translate import Translate
from pythainlp.tokenize import sent_tokenize
th2en = Translate('th', 'en')
en2th = Translate('en', 'th')
```

#### Sentence Segmentation

```python
def printlen(w):
  print(f'len: {len(w)}')
  print(w)

def join_text_range(article_sent, range_select):
  return ''.join(article_sent[0:range_select]) 

def cut_sent_to_fit(article_text):
  article_sent = sent_tokenize(article_text, engine="crfcut")
  if article_text <= 1000:
    return article_sent
  
  else:
    iloc_ = 14 # Start location
    while True:
      article_cut = join_text_range(article_sent, iloc_)
      len_article = len(article_cut)
      if len_article >= 1000:
        iloc_ -= 1
      elif len_article <= 900:
        iloc_ += 1
      else:
        # break when len_article range between 900 and 1000
        break
  print(f"iloc: {iloc_}")
  return join_text_range(article_sent, iloc_)
```

```python
article_text = df['body'][1]
article_text = """
 3. ตอนอยู่ ม. 4 (ปี 1996) ได้อ่านหนังสือ “โครงสร้างข้อมูล: เพื่อการออกแบบโปรแกรมคอมพิวเตอร์” ของ ดร.สุชาย ธนวเสถียร และ อ.วิชัย จิวังกูร 
ผลคือ: ทำให้รู้ว่า คณิตศาสตร์ที่เปลี่ยนเป็นภาพได้ เขียนโปรแกรมพิสูจน์ได้ และใช้งานได้จริง มันเป็นแบบนี้นี่เอง พอวิชา data structures แน่น ก็เลยทำให้เรียนวิชา algorithms ได้แบบชิลมาก (ใช้หนังสือ Algortihms ของ Thomas H. Cormen และคณะ)
สภาพปัจจุบัน: ชอบเปลี่ยนไอเดียทางคณิตศาสตร์ให้กลายเป็นภาพ และพยายามหาคำอธิบายที่ intuitive ก่อน แล้วค่อยนำสิ่งนี้ไปถ่ายทอดให้คนอื่นต่ออีกที
"""

# article_small_text = cut_sent_to_fit(article_text)
article_small_text = (article_text)
printlen(article_small_text)
```

```python
article_text_en = th2en.translate(article_small_text)
printlen(article_text_en)
```

```python
import torch
from transformers import AutoTokenizer, AutoModelWithLMHead

tokenizer = AutoTokenizer.from_pretrained("deep-learning-analytics/wikihow-t5-small")
model = AutoModelWithLMHead.from_pretrained("deep-learning-analytics/wikihow-t5-small")

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = model.to(device)

# ARTICLE HERE
text = article_text_en

preprocess_text = text.strip().replace("\n","")
tokenized_text = tokenizer.encode(preprocess_text, return_tensors="pt").to(device)

summary_ids = model.generate(
            tokenized_text,
            max_length=150, 
            num_beams=2,
            repetition_penalty=2.5, 
            length_penalty=1.0, 
            early_stopping=True
        )


summary_eng = (tokenizer.decode(summary_ids[0], skip_special_tokens=True))
summary = en2th.translate(summary_eng)

print ("\n\nSummarized text: \n",summary)
```
