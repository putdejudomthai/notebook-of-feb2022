```python
import re, time
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from pythainlp import sent_tokenize
import torch
device = "cuda:0" if torch.cuda.is_available() else "cpu"

def join_text_range(article_sent, range_select):
  return ''.join(article_sent[0:range_select]) 

def cut_sent_to_fit(article_text):
  article_sent = sent_tokenize(article_text, engine="crfcut")
  if len(article_text) <= 700:
    return article_text
  
  else:
    iloc_ = 9 # Start location
    while True:
      article_cut = join_text_range(article_sent, iloc_)
      len_article = len(article_cut)
      if len_article >= 700:
        iloc_ -= 1
      elif len_article <= 600:
        iloc_ += 1
      else:
        # break when len_article range between 900 and 1000
        break
  #print(f"iloc: {iloc_}")
  return join_text_range(article_sent, iloc_)
class lnwharry():
    def __init__(self):
        # Init to load model
        self.WHITESPACE_HANDLER = lambda k: re.sub('\s+', ' ', re.sub('\n+', ' ', k.strip()))
        model_name = "csebuetnlp/mT5_multilingual_XLSum"
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

    def summary(self, article_text,
                max_length_input = 512,
                max_length_output = 84,
                no_repeat_ngram_size = 2,
                num_beams = 4,
                ):
        """
        Input: long article text
        Output: summary text
        """

        #article_text = cut_sent_to_fit(article_text)

        input_ids = self.tokenizer(
            [self.WHITESPACE_HANDLER(article_text)],
            return_tensors="pt",
            padding = "max_length",
            truncation = True,
            #max_length = 512
            max_length = max_length_input # OG
        )["input_ids"].to(device)
        self.model.to(device)
        output_ids = self.model.generate(
            input_ids=input_ids,
            max_length = max_length_output, # OG
            # max_length=200,

            no_repeat_ngram_size=no_repeat_ngram_size,
            # no_repeat_ngram_size = 3 is not repeat 3-gram

            num_beams=num_beams
        )[0]

        summary = self.tokenizer.decode(
            output_ids,
            skip_special_tokens=True,
            clean_up_tokenization_spaces=False
        )
        return summary
```
