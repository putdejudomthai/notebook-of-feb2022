```python
import os
train_image_path = []
train_json_path = []
for root, dirs, files in os.walk(train_dir_path):
    for name in files:
        if name.endswith('.png'):
            train_image_path.append(os.path.join(root, name))
        if name.endswith('.json'):
            train_json_path.append(os.path.join(root, name))
```

```python
from tqdm import tqdm
```

```python
import cv2
import numpy
import json
import random

case_error = []
output_fonts = ['Angsana_New', 'Cordia_New', 'DM_Shining_Star_Regular', 'FC_Knomphing_Regular', 
                'fonttintin', 'Kunlasatri', 'TH_Chakra_Petch', 'TH_Charm_of_AU', 'TH_Mali_Grade6', 'TH_Sarabun']
output_size = ['12px/','14px/','16px/','18px/','20px/','22px/','24px/']
output_style = ['normal-normal/','normal-bold/','italic-normal/','italic-bold/']

output_path = 'SuperTestSet/'
for path in output_fonts+output_size+output_style: 
    os.makedirs(output_path + path, exist_ok = True)

def crop_image(start, end):
    for index in tqdm(range(start, end)):
        # Read image and json file.
        name = '-'.join(train_image_path[index].split('/')[-3:-1])
        image = cv2.imread(train_image_path[index])
        f = open (train_json_path[index], "r")
        image_dict = json.loads(f.read())

        # Bounding box and Save
        for i, info in enumerate(image_dict):
            # Crop Image (Trust bounding box)
            # x, y, w, h = map(int, info["rect"].values())
            x = int(info["x"])
            y = int(info["y"])
            w = int(info["width"])
            h = int(info["height"])
            # EDIT
            crop_image = image[y-1:y+h+2, x-1:x+w+2]
            # fam, size, style, _, weight = info["style"].values()
            # size = size[:-2]
            # id = info["id"]

            # parentId = info["parentId"]
            parentId = info["id"]

            if crop_image.shape[0] != 0 and crop_image.shape[1] != 0:
                # EDIT FILE NAME
                # cv2.imwrite(f'{output_path}{fam}/{fam}-{size}-{style}-{weight}-{parentId}.jpg', crop_image)
                # cv2.imwrite(f'{output_path}{size}px/{fam}-{size}-{style}-{weight}-{parentId}.jpg', crop_image)
                # cv2.imwrite(f'{output_path}{style}-{weight}/{fam}-{size}-{style}-{weight}-{parentId}.jpg', crop_image)
                cv2.imwrite(f'{output_path}{parentId}.jpg', crop_image)
```
