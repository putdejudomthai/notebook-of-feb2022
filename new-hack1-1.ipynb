```python
# expand image to have same height
def fix_h(array):
    if array.shape[0]<height:
        img_pad = cv2.resize(array, (int(height*array.shape[1]/array.shape[0]), height) )  # adjust every image based on height
    else:
        img_pad = array[:height,:]  #cut
    return img_pad
```

```python
# expand image to have same width
# have chance to cut off some text
# for example the first picture the width is 123 -> but we want to increase it to 340 as requirement
def padding(array):
    if array.shape[1]<width:
        pad_left = int( (width-array.shape[1])/2 )
        pad_right = width-pad_left-array.shape[1]
        img_pad = cv2.copyMakeBorder(array, 0,0, pad_left, pad_right, cv2.BORDER_CONSTANT, 255)
    else:
        img_pad = array[:,:width]
    return img_pad
```

```python
def empty_check(array):
    area = np.sum(array)
    return area<1000  ### พื้นหลัง เปนสีดำ มีแต่ 0 ก็คือน้อยมากๆ อาจจะเปนจุด
```

```python
def prepare_img(path):
    f = '../input/w1n1-data-v2/crop_images'+'/'+path
    img = cv2.imread(f,cv2.IMREAD_GRAYSCALE) 
    invert = (255-img)
    if empty_check(invert):
        return 0
    else:
        return padding(fix_h(invert))
```

```python
img_list = []
for f in data['file']:
    img_list.append(prepare_img(f))
img_list = np.array(img_list)
```
