```python
class corpus():
  def dow(self,name):
    download(name)
    path = get_corpus_path(name)
    cor = get_corpus(path)
    word = ["".join(x.split("\t")[:-1]) for x in cor]
    fre = ["".join(x.split("\t")[-1]) for x in cor]
    dict_cor = {"word":word,"fre":fre}
    self.df = pd.DataFrame(dict_cor,index=range(len(word))).set_index('word')
    self.path = path
    self.word = word
    self.fre = fre
  
  def __init__(self,name):
    self.dow(name)

  def __len__(self):
    return self.df.shape[0]

  def __getitem__(self,word):
    return self.df.loc[word]
```

```python
def edits1(word):
    "All edits that are one edit away from `word`."
    letters    = 'กขฃคฅฆงจฉชซฌญฎฐฑฒณดตถทธนบปผฝพฟภมยรลวศษสหฬอฮ'
    splits     = [(word[:i], word[i:])    for i in range(len(word) + 1)]
    deletes    = [L + R[1:]               for L, R in splits if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R)>1]
    replaces   = [L + c + R[1:]           for L, R in splits if R for c in letters]
    inserts    = [L + c + R               for L, R in splits for c in letters]
    return set(deletes + transposes + replaces + inserts)
```
